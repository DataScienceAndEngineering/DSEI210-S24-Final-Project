**To Clean the columns 'VEHICLE TYPE CODE 1'and 'VEHICLE TYPE CODE 2'**

There are some vehicle type names which are named differently in the dataset. So we have to rename them properly.

```python
print(data['VEHICLE TYPE CODE 1'])
```

```python
unique_vehicle_type = data['VEHICLE TYPE CODE 1'].nunique()
print(unique_vehicle_type)
```

```python
data['VEHICLE TYPE CODE 1'] = data['VEHICLE TYPE CODE 1'].str.upper()
vehicle_type = data['VEHICLE TYPE CODE 1'].nunique()
print(vehicle_type)
```

```python
unique_vehicle_type2 = data['VEHICLE TYPE CODE 2'].nunique()
print(unique_vehicle_type)
```

```python
data['VEHICLE TYPE CODE 2'] = data['VEHICLE TYPE CODE 2'].str.upper()
vehicle_type2 = data['VEHICLE TYPE CODE 2'].nunique()
print(vehicle_type2)
```

```python
data.head()
```

```python
total_vehicle_types = pd.concat([data['VEHICLE TYPE CODE 1'], data['VEHICLE TYPE CODE 2']])
```

```python
tot_vehicle_types = total_vehicle_types.unique()
print(tot_vehicle_types)
```

```python
total_vehicle_types.dropna(inplace=True)
type_counts = total_vehicle_types.value_counts()
print(type_counts.head(50))

```

```python
vehicle_type_map = {
    'STATION WAGON/SPORT UTILITY VEHICLE': 'SUV',
    'SPORT UTILITY / STATION WAGON': 'SUV',
    '4 DR SEDAN': 'SEDAN',
    '2 DR SEDAN': 'SEDAN',
    'PASSENGER VEHICLE': 'SEDAN',
    'SMALL COM VEH(4 TIRES) ': 'SMALL COMMERCIAL VEHICLE',
    'SMALL COM VEH(4 TIRES)': 'SMALL COMMERCIAL VEHICLE',
    'LARGE COM VEH(6 OR MORE TIRES)': 'LARGE COMMERCIAL VEHICLE',
    'TRACTOR TRUCK DIESEL': 'TRACTOR TRUCK',
    'TRACTOR TRUCK GASOLINE': 'TRACTOR TRUCK',
    'MOTORSCOOTER': 'SCOOTER',
    'MOTORBIKE': 'MOTORCYCLE',
    'E-BIKE': 'ELECTRIC BIKE',
    'E-SCOOTER': 'ELECTRIC SCOOTER',
    'AMBUL': 'AMBULANCE',
    'FIRE' : 'FIRE TRUCK',
    'GARBAGE OR REFUSE' : 'GARBAGE TRUCK',
    'E-BIK' : 'ELECTRIC BIKE',
    'E-SCO' : 'ELECTRIC SCOOTER',
    'OTHER' : 'UNKNOWN'
}
```

```python
total_vehicle_types.replace(vehicle_type_map, inplace=True)
type_counts = total_vehicle_types.value_counts()
print(type_counts.head(50))
```
